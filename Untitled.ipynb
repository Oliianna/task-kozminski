```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

marketing = pd.read_csv('marketing.csv', delimiter=';')
marketing
```

```python
#Adding new columns: the CR(conversion rate, %) and CPA (cost per acquisition), CPC (cost per click)

marketing['CR,%'] = (marketing['Conversions']/marketing['Sessions'])*100
marketing['CPA'] = round(marketing['Marketing_budget']/marketing['Conversions'], 2)
marketing['CPC'] = round(marketing['Marketing_budget']/marketing['Sessions'], 2)
marketing
```

```python
print(marketing.sum())
```

```python
#Bar chart of sessions by channel

marketing['Sessions'].plot(kind = 'bar', color = "#A17BFB")
plt.title('Number of sessions by channel')
plt.ylabel('Sessions')
plt.xticks(range(9), marketing['Channel'])
plt.show()
```

```python
#Pie chart of sessions by channel

colorPalette = ["#FAECDE", "#A17BFB","#D5C5FA","#AA9DC7","#ACFAB6", "#80AD86", "#66AD6F", "#AD8E6F", "#614B34"]
plt.pie(marketing['Sessions'], labels = marketing['Channel'], radius = 1.7, colors = colorPalette)
plt.show()
```

```python
#Bar chart of CR by channels

marketing['CR,%'].plot(kind = 'bar', color = "#A17BFB")
plt.title('Cr, % by channel')
plt.ylabel('CR,%')
plt.xticks(range(9), marketing['Channel'])
plt.show()
```
